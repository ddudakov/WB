Так как nm_id идут не подряд, в том смысле, что в head выше мы видим nm_id 399340, когда уникальных всего - 386499, с юзерами присутствует такая же проблема, то лучше всего закодировать по порядку nm и user, напишем функцию

```python
from sklearn.preprocessing import LabelEncoder

def id_encoder(df):
    ds = df.copy()
    users = sorted(ds["wbuser_id"].unique())
    items = sorted(ds["nm_id"].unique())

    # Создаем енкодеры для пользователей и товаров
    uencoder = LabelEncoder()
    iencoder = LabelEncoder()
    
    uencoder.fit(users)
    iencoder.fit(items)

    # соответственно кодируем
    ds.wbuser_id = uencoder.transform(ds.wbuser_id.tolist())
    ds.nm_id = iencoder.transform(ds.nm_id.tolist())

    return ds, uencoder, iencoder
```

## Функция для разбиения датасета по времени

```python
def train_test_split(df, quant = 0.85):
    threshold = df['dt'].quantile(q=quant)
    df_train = df[df['dt'] < threshold]
    df_test = df[df['dt'] >= threshold]
    return df_train, df_test
#Всего получится 5 датафреймов, трейн/вал для кандидатов, трейн/вал для ранжирующей и тест
```

### У нас есть матрица взаимодействий(с помощью coo_matrix), теперь нужна реализовать функцию, которая поможет производить валидацию

#### Для этого нужно выделить сессии в валидационной части, будем считать, что пары item-item где один айтем фиксированный, а другой из его сессии - за лейблы, тем самым проверять будем метрикой так, чтобы предсказанные айтемы для рекомендации, попадали в множество лейблов 

#### Нам нужно добавить сессии, предположим, что если пользователь не взаимодействовал с товарами больше 15 минут, относительно последнего взаимодействия, его дальнейшие взаимодействия будут считаться за новую сессию

```python
def session_id(df):
    df = df.sort_values(by=["wbuser_id", "dt"]).reset_index(drop=True)
    df["time_diff"] = df.groupby("wbuser_id")["dt"].diff().dt.total_seconds()
    df["is_new_session"] = (df["wbuser_id"] != df["wbuser_id"].shift(1)) | (df["time_diff"] > 15 * 60)
    df["session_id"] = df["is_new_session"].cumsum() - 1
    df = df.drop(columns=["time_diff", "is_new_session"])
    return df

ds_cand_with_session = session_id(ds_cand_train_new_id)
print(ds_cand_with_session)
```
