## Функция для разбиения датасета по времени

```python
def train_test_split(df, quant = 0.85):
    threshold = df['dt'].quantile(q=quant)
    df_train = df[df['dt'] < threshold]
    df_test = df[df['dt'] >= threshold]
    return df_train, df_test
#Всего получится 5 датафреймов, трейн/вал для кандидатов, трейн/вал для ранжирующей и тест
```

```python
df_train_all, df_test = train_test_split(df_click)
df_cand, df_range = train_test_split(df_train_all, quant = 0.5)
df_cand_train, df_cand_val = train_test_split(df_cand, quant = 0.75)
df_range_train, df_range_val = train_test_split(df_range, quant = 0.75)
print(df_cand_train, df_cand_train['nm_id'].nunique(), df_cand_train['nm_id'].max())
```

# Для кандидатов

Так как nm_id и wbuser_id идут не подряд, то лучше всего закодировать по порядку nm и user, так как это пригодится нам для составления user-item матрицы, которые требуют некоторые кандидаты, напишем функцию

```python
from sklearn.preprocessing import LabelEncoder

def id_encoder(df):
    ds = df.copy()
    users = sorted(ds["wbuser_id"].unique())
    items = sorted(ds["nm_id"].unique())

    # Создаем енкодеры для пользователей и товаров
    uencoder = LabelEncoder()
    iencoder = LabelEncoder()
    
    uencoder.fit(users)
    iencoder.fit(items)

    # соответственно кодируем
    ds.wbuser_id = uencoder.transform(ds.wbuser_id.tolist())
    ds.nm_id = iencoder.transform(ds.nm_id.tolist())

    return ds, uencoder, iencoder
```

### Теперь у нас пронумерованы без "пробелов" id, можем заняться матрицей user-item

```python
import numpy as np
from scipy.sparse import coo_matrix

def matrix_user_item(df, user_enc, item_enc):
    user_ids = df["wbuser_id"]
    item_ids = df["nm_id"]
    values = [1] * len(df)
    interaction_matrix = coo_matrix((values, (user_ids, item_ids)), shape=(len(user_enc.classes_), len(item_enc.classes_)))
    return interaction_matrix.toarray()

mat_for_cand = matrix_user_item(df_cand_train_new_id, ue_cand, ie_cand)
print(mat_for_cand)
```

#### Чтобы выполнить валидацию нужно выделить сессии в валидационной части, будем считать, что пары item-item где один айтем фиксированный, а другой из его сессии - за лейблы, тем самым проверять будем метрикой так, чтобы предсказанные айтемы для фиксированного, попадали в множество лейблов 

#### Нам нужно добавить сессии, предположим, что если пользователь не взаимодействовал с товарами больше 15 минут, относительно последнего взаимодействия, его дальнейшие взаимодействия будут считаться за новую сессию

```python
def session_id(df):
    df = df.sort_values(by=["wbuser_id", "dt"]).reset_index(drop=True)
    df["time_diff"] = df.groupby("wbuser_id")["dt"].diff().dt.total_seconds()
    df["is_new_session"] = (df["wbuser_id"] != df["wbuser_id"].shift(1)) | (df["time_diff"] > 15 * 60)
    df["session_id"] = df["is_new_session"].cumsum() - 1
    df = df.drop(columns=["time_diff", "is_new_session"])
    return df

df_cand_val_with_session = session_id(df_cand_val_new_id)
print(df_cand_val_with_session[:50])
# df_cand_train_with_session = session_id(df_cand_train_new_id)
# print(ds_cand_with_session)
```

### Получили сессии по нашему условию, теперь нужно преобразовать их просто в набор айтемов и взять по парам, за true_label будем считать встречу двух айтемов в одной сессии, также посчитаем количество каждой такой пары

```python
print(df_cand_val_with_session.groupby('session_id')['nm_id'].apply(list))
```

```python
from itertools import combinations
from collections import Counter

def generate_pairs_with_counts(df):
    df_with_session = session_id(df)
    # Группируем товары по session_id
    item_groups = df_with_session.groupby('session_id')['nm_id'].apply(list)
    item_groups = item_groups[item_groups.apply(len) > 1]  # Убираем сессии с одним товаром
    # Счетчик для хранения количества повторений пар
    pair_counts = Counter()
    for items in item_groups:
        pairs = combinations(sorted(items), 2)  # Создаем пары и сортируем для уникальности
        pair_counts.update(pairs)

    # Преобразуем счетчики в DataFrame
    pairs_df = pd.DataFrame([
        {'item_anchor': pair[0], 'item': pair[1], 'count': count}
        for pair, count in pair_counts.items() if count > 10 # это можно задать гиперпараметром, и выбрать число, оно означает количество сессий в которых встретились эти пары
    ])
    
    return pairs_df

df_cand_val_pairs = generate_pairs_with_counts(df_cand_val_new_id)
print(df_cand_val_pairs)
```

### Нам нужно нагенерировать негативных примеров, для этого мы воспользуемся функцией generate_pairs_with_counts которая превратит наш датафрейм в пары и после возьмем за негативы, те примеры, которые не были в парах

```python
# Функция для генерации негативных примеров
def generate_negative_samples(df, all_items, num_negatives = 3):
    df = df.drop(columns=['count'])
    negative_samples = []
    # Для каждого якорного товара и товара из сессии
    for index, row in df.iterrows():
        anchor_item = row['item_anchor']
        item = row['item']
        negative_items = list(all_items.difference(set([anchor_item, item])))
        for _ in range(num_negatives):
            # Выбираем случайный негативный товар
            negative_item = np.random.choice(negative_items)
            negative_samples.append([anchor_item, negative_item])
    negative_df = pd.DataFrame(negative_samples, columns=['item_anchor', 'item'])
    return negative_df

def validate_for_range(df, all_items):
    negative_samples_df = generate_negative_samples(df, all_items)
    df['click'] = 1
    negative_samples_df['click'] = 0
    df_all = pd.concat([df, negative_samples_df], ignore_index=True)
    return df_all
```
